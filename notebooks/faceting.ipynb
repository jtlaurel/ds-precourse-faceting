### Challenge 2

For this challenge you will be working with the **cars** data set. Run the cell below to import the data set.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cars_data = pd.read_csv('../data/cars.csv', sep='|')
cars_data.rename(columns={'Unnamed: 0': 'make_model'}, inplace=True)
cars_data.head()
````

For this challenge:
* Facet the plots on the number of cylenders each vechical has
* Create a single row with with the number of correct plots
* Use a scatter plot to show the relation between horse power and quarter mile time
* On the x axis plot the horse power
* On the y axis plot the the quarter mile time (qsec)
* Each plot should have a title saying the number of cylinders
* Give each axis an appropriate name

````python
# Write your code here
cylinders = cars_data.groupby(['cyl'])
fourcyl = cylinders.get_group(4)
sixcyl = cylinders.get_group(6)
eightcyl = cylinders.get_group(8)

plt.figure()
plt.figtext(0.5, 0.95, "Horse Power vs Quarter Mile Time for Different Cylinder Cars", ha="center", fontsize=10)

plt.subplot(1,3,1)
plt.scatter(fourcyl['hp'], fourcyl['qsec'])
plt.ylim([13,24])
plt.yticks(range(13,24,2))
plt.title("Four Cylinder")
plt.xlabel("Horsepower")
plt.ylabel("Quarter Mile Time (in sec)")

plt.subplot(1,3,2)
plt.scatter(sixcyl['hp'], sixcyl['qsec'])
plt.ylim([13,24])
plt.yticks([])
plt.title("Six Cylinder")
plt.xlabel("Horsepower")

plt.subplot(1,3,3)
plt.scatter(eightcyl['hp'], eightcyl['qsec'])
plt.ylim([13,24])
plt.yticks([])
plt.title("Eight Cylinder")
plt.xlabel("Horsepower")

plt.show()
````
